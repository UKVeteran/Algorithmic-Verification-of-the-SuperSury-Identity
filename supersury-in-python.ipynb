```python
from fractions import Fraction

def get_fibonacci_lucas(max_n):
    F, L = [0, 1], [2, 1]
    for _ in range(2, max_n + 3):
        F.append(F[-1] + F[-2])
        L.append(L[-1] + L[-2])
    return F, L

def get_a015530(max_n):
    a = [0, 1]
    for _ in range(2, max_n + 3):
        a.append(4 * a[-1] + 3 * a[-2])
    return a

def get_custom_recurrence(x0, x1, c1, c2, max_n):
    """Generates an arbitrary second-order recurrence sequence."""
    X = [x0, x1]
    for _ in range(2, max_n + 3):
        X.append(c1 * X[-1] + c2 * X[-2])
    return X

def verify_identities(n_range=8):
    F, L = get_fibonacci_lucas(3 * n_range + 5)
    a = get_a015530(n_range + 5)
    
    # 1. Output original requested verifications
    print("=== VERIFYING DRESDEN & GAO (2026) SUMMATIONS ===")
    print(f"{'n':<5} | {'Identity 1 (Lucas) LHS':<25} | {'Identity 1 RHS':<25} | {'Match':<5}")
    print("-" * 68)
    for n in range(n_range):
        lhs = sum((-2)**i * L[3*i] for i in range(n + 1))
        rhs = 2 - (-2)**(n + 1) * F[3*n]
        print(f"{n:<5} | {lhs:<25} | {rhs:<25} | {str(lhs == rhs):<5}")
        
    print("\n" + "="*68 + "\n")
    print(f"{'n':<5} | {'Identity 2 (A015530) LHS':<25} | {'Identity 2 RHS':<25} | {'Match':<5}")
    print("-" * 68)
    for n in range(n_range):
        lhs = Fraction(a[n + 1], 1)
        weighted_sum = sum(Fraction(-3, 4)**(n - i) * Fraction(a[i + 2], 1) for i in range(n + 1))
        rhs = Fraction(1, 4) * weighted_sum
        print(f"{n:<5} | {str(lhs):<25} | {str(rhs):<25} | {str(lhs == rhs):<5}")

    # 2. General SuperSury Identity Verification (Theorem 2 Master Formula)
    print("\n" + "="*68 + "\n")
    print("=== VERIFYING THE GENERAL SUPERSURY IDENTITY (THEOREM 2) ===")
    
    # Defining an entirely arbitrary sequence: X_n = 5*X_{n-1} - 2*X_{n-2}
    x0, x1, c1, c2, k = 3, 7, 5, -2, 2
    X = get_custom_recurrence(x0, x1, c1, c2, n_range + 5)
    
    # Compute dynamically calculated parameters from Theorem 2
    # t = -c2 * X_{k-1} / X_k
    t = Fraction(-c2 * X[k-1], X[k])
    # scaling factor = (X_0 * X_2 - X_1^2) / X_k
    scaling_factor = Fraction(X[0] * X[2] - X[1]**2, X[k])
    
    print(f"Sequence configuration: X_0={x0}, X_1={x1}, c1={c1}, c2={c2}, k={k}")
    print(f"Computed Target Weight (t) = {t}")
    print(f"Computed Scale Front Factor = {scaling_factor}\n")
    print(f"{'n':<5} | {'SuperSury LHS':<25} | {'SuperSury RHS':<25} | {'Match':<5}")
    print("-" * 68)
    
    for n in range(n_range):
        # LHS: X_0 * X_{n+2} - X_1 * X_{n+1}
        super_lhs = Fraction(X[0] * X[n + 2] - X[1] * X[n + 1], 1)
        
        # RHS: ScalingFactor * Sum_{i=0}^n (t)^{n-i} * X_{i+k}
        super_sum = sum(t**(n - i) * Fraction(X[i + k], 1) for i in range(n + 1))
        super_rhs = scaling_factor * super_sum
        
        print(f"{n:<5} | {str(super_lhs):<25} | {str(super_rhs):<25} | {str(super_lhs == super_rhs):<5}")

if __name__ == "__main__":
    verify_identities(n_range=8)
```
